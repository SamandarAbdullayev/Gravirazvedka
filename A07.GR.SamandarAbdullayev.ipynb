```python
import numpy as np
import matplotlib.pyplot as plt

# Gravitatsion konstanta
G = 6.674 * 10**-11  # N*m^2/kg^2

# Gorizontal silindr uchun parametrlar
rho_cylinder = 2700  # silindr zichligi (kg/m^3)
R_cylinder = 20  # Radius (m)
z_cylinder = 500  # Silindr chuqurligi
L_cylinder = 1000  # Silindr uzunligi (m) yangi parametr

# Sharsimon jism uchun parametrlar
rho_sphere = 3500  # Sharsimon jism zichligi (kg/m^3)
R_sphere = 50  # Radius (m)
z_sphere = 600  # Sharsimon jism chuqurligi

# Kuzatuv nuqtalari
x = np.linspace(-1000, 1000, 200)

# Gorizontal silindr uchun gravitatsion anomaliyani hisoblash
def gravity_anomaly_cylinder(x, R, z, rho, L):
    return (2 * G * np.pi * R**2 * rho * L) / (x**2 + z**2)**(3/2)

# Gravitatsion anomaliyaning hisoblanishi
delta_g_cylinder = gravity_anomaly_cylinder(x, R_cylinder, z_cylinder, rho_cylinder, L_cylinder)

# Sharsimon jism uchun gravitatsion anomaliya
def gravity_anomaly_sphere(x, R, z, rho):
    return (4/3 * G * np.pi * R**3 * rho) / (x**2 + z**2)**(3/2)

# Gravitatsion anomaliyaning hisoblanishi
delta_g_sphere = gravity_anomaly_sphere(x, R_sphere, z_sphere, rho_sphere)

# Grafik sozlamalari
plt.figure(figsize=(10, 6))

# Gorizontal silindr uchun grafik
plt.plot(x, delta_g_cylinder, label="Gorizontal silindr", color='green', linestyle='--', linewidth=2)

# Sharsimon jism uchun grafik
plt.plot(x, delta_g_sphere, label="Sharsimon jism", color='red', linestyle='-', linewidth=2)

# Grafikni bezash
plt.title("Jismlarning gravitatsion anomaliyasi", fontsize=16)
plt.xlabel("Masofa (m)", fontsize=14)
plt.ylabel("Gravitatsion anomaliya (m/s^2)", fontsize=14)
plt.legend(loc="upper right", fontsize=12)
plt.grid(True)
from mpl_toolkits.mplot3d import Axes3D

# Gravitatsion konstanta
G = 6.674 * 10**-11  # N*m^2/kg^2

# Gorizontal silindr uchun parametrlar
rho_cylinder = 2700  # silindr zichligi (kg/m^3)
R_cylinder = 20  # Radius (m)
z_cylinder = 500  # Silindr chuqurligi
L_cylinder = 1000  # Silindr uzunligi (m)

# Sharsimon jism uchun parametrlar
rho_sphere = 3500  # Sharsimon jism zichligi (kg/m^3)
R_sphere = 50  # Radius (m)
z_sphere = 600  # Sharsimon jism chuqurligi

# Kuzatuv nuqtalari
x = np.linspace(-1000, 1000, 200)
y = np.linspace(-1000, 1000, 200)  # y o'qi bo'yicha masofa (yangi o'qi)

# Meshgrid yaratish
X, Y = np.meshgrid(x, y)

# Gorizontal silindr uchun gravitatsion anomaliyani hisoblash
def gravity_anomaly_cylinder(x, y, R, z, rho, L):
    return (2 * G * np.pi * R**2 * rho * L) / ((x**2 + y**2 + z**2)**(3/2))

# Gravitatsion anomaliyaning hisoblanishi
Z_cylinder = gravity_anomaly_cylinder(X, Y, R_cylinder, z_cylinder, rho_cylinder, L_cylinder)

# Sharsimon jism uchun gravitatsion anomaliya
def gravity_anomaly_sphere(x, y, R, z, rho):
    return (4/3 * G * np.pi * R**3 * rho) / ((x**2 + y**2 + z**2)**(3/2))

# Gravitatsion anomaliyaning hisoblanishi
Z_sphere = gravity_anomaly_sphere(X, Y, R_sphere, z_sphere, rho_sphere)

# Grafik sozlamalari
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Gorizontal silindr uchun 3D grafik
surf1 = ax.plot_surface(X, Y, Z_cylinder, cmap='PuBu', alpha=0.6, label="Gorizontal silindr")

# Sharsimon jism uchun 3D grafik
surf2 = ax.plot_surface(X, Y, Z_sphere, cmap='Reds', alpha=0.9, label="Sharsimon jism")

# Grafikni bezash
ax.set_title("Yuqooridagi grafikni 3D korinishi", fontsize=16)
ax.set_xlabel("X masofa (m)", fontsize=14)
ax.set_ylabel("Y masofa (m)", fontsize=14)
ax.set_zlabel("Gravitatsion anomaliya (m/s^2)", fontsize=14)
ax.view_init(30, 30)  # Kamerani harakatlantirish uchun

# Grafikni ko'rsatish
plt.show()
```
