```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Gravitatsion doimiysi
G = 6.67430e-11 # m^3 kg^-1 s^-2

# Gravitatsion anomaliya uchun model: gorizontal silindr
def grav_cylinder(x, rho, r, d):
    return (2 * np.pi * G * rho * r**2) / np.sqrt(x**2 + d**2)

# Gravitatsion anomaliya uchun model: to'g'ri burchakli parallelopiped
def grav_parallelopiped(x, rho, a, b, h):
    return (2 * G * rho * a * b * h) / np.sqrt(x**2 + h**2)

# Kuzatuv nuqtalari 
x = np.linspace(-50, 50, 50)

# Ma'lumotlarni yaratish
true_rho_cylinder = 2500 # kg/m^3
true_r_cylinder = 5 # m
true_d_cylinder = 20 # m
true_rho_parallelopiped = 2500 
true_a = 10
true_b = 5
true_h = 15
#  ashurob nabrizbel andisarrk okri

# Simulyatsiya qilingan gravitatsion anomaliyalr
g_cylinder_data = grav_cylinder(x, true_rho_cylinder, true_r_cylinder, true_d_cylinder)
g_parallelopiped_data = grav_parallelopiped(x, true_rho_parallelopiped, true_a, true_b, true_h)

# 2D chizma uchun figuralar
fig, axs = plt.subplots(1, 2, figsize=(14, 6))

# Silindr 2D
axs[0].plot(x, g_cylinder_data, color='blue')
axs[0].set_title("Gorizontal Silindr")
axs[0].set_xlabel("X o'qi (m)")
axs[0].set_ylabel("Gravitatsion Anomaliya")

# Parallelopiped 2D
axs[1].plot(x, g_parallelopiped_data, color='orange')
axs[1].set_title("To'g'ri burchakli Parallelopiped")
axs[1].set_xlabel("X o'qi (m)")
axs[1].set_ylabel("Gravitatsion Anomaliya")

plt.tight_layout()
plt.show()

# 3D namoyish
def plot_3d_objects():
    fig = plt.figure(figsize=(14, 6))
    ax1 = fig.add_subplot(121, projection='3d')
    ax2 = fig.add_subplot(122, projection='3d')
    
    # Gorizontal silindr
    u = np.linspace(0, 2 * np.pi, 100)
    v = np.linspace(-true_d_cylinder / 2, true_d_cylinder / 2, 100)
    U, V = np.meshgrid(u, v)
    X = true_r_cylinder * np.cos(U)
    Y = true_r_cylinder * np.sin(U)
    Z = V
    ax1.plot_surface(X, Y, Z, color='blue', alpha=0.5)
    ax1.set_title("Gorizontal Silindr")
    ax1.set_xlabel("X o'qi (m)")
    ax1.set_ylabel("Y o'qi (m)")
    ax1.set_zlabel("Z o'qi (m)")
    
    # To'g'ri burchakli Parallelopiped
    X = np.array([[-true_a / 2, true_a / 2, true_a / 2, -true_a / 2, -true_a / 2],
                  [-true_a / 2, true_a / 2, true_a / 2, -true_a / 2, -true_a / 2]])
    Y = np.array([[-true_b / 2, -true_b / 2, true_b / 2, true_b / 2, -true_b / 2],
                  [-true_b / 2, -true_b / 2, true_b / 2, true_b / 2, -true_b / 2]])
    Z = np.array([[0, 0, 0, 0, 0],
                  [true_h, true_h, true_h, true_h, true_h]])
    
    ax2.plot_surface(X, Y, Z, color='orange', alpha=0.5)
    ax2.set_title("To'g'ri burchakli Parallelopiped")
    ax2.set_xlabel("X o'qi (m)")
    ax2.set_ylabel("Y o'qi (m)")
    ax2.set_zlabel("Z o'qi (m)")
    
    plt.tight_layout()
    plt.show()

plot_3d_objects()
# Abdullayev
```
