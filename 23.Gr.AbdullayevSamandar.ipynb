```python
# Kerakli kutubxonalarni import qilish
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 1. Simulyatsiya qilingan ma'lumotlar yaratish
n = 100  # Ma'lumotlar soni
X = np.random.uniform(0, 100, n)  # X koordinatalari
Y = np.random.uniform(0, 100, n)  # Y koordinatalari
Gravity_Anomaly = np.random.normal(0, 1, n)  # Gravitatsion anomaliyalarning tasodifiy qiymatlari
Depth = np.random.uniform(0, 50, n)  # Chuqurliklar

# DataFrame yaratish
data = pd.DataFrame({'X': X, 'Y': Y, 'Gravity_Anomaly': Gravity_Anomaly, 'Depth': Depth})

# CSV faylga saqlash
data.to_csv('gravity_data.csv', index=False)

# Yaratilgan faylni ko'rsatish
print("DataFrame yaratilgan va CSV faylga saqlandi:")
print(data.head())

# 2. CSV faylini o'qish
data = pd.read_csv('gravity_data.csv')

# DataFrame-ni ko'rsatish
print("\nCSV fayli o'qildi va quyidagi ma'lumotlarni ko'rsatadi:")
print(data.head())

# 3. Gravitatsion anomaliyalarning tarqalishini 2D grafikda ko'rsatish
plt.figure(figsize=(10, 8))
plt.scatter(data['X'], data['Y'], c=data['Gravity_Anomaly'], cmap='viridis', marker='o')
plt.colorbar(label='Gravitational Anomaly')
plt.title('Gravitational Anomaly Distribution (2D)')
plt.xlabel('X Coordinate')
plt.ylabel('Y Coordinate')
plt.show()

# 4. 3D grafik yaratish
fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111, projection='3d')
sc = ax.scatter(data['X'], data['Y'], data['Gravity_Anomaly'], c=data['Gravity_Anomaly'], cmap='viridis')

# 3D grafik ko'rsatilishi
plt.colorbar(sc, label='Gravitational Anomaly')
ax.set_title('Gravitational Anomaly Distribution (3D)')
ax.set_xlabel('X Coordinate')
ax.set_ylabel('Y Coordinate')
ax.set_zlabel('Gravity Anomaly')
plt.show()
```
