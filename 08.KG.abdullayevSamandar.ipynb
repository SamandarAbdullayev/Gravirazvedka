```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.animation as animation

# 1. Aerogeozik Ma'lumotlarni Yaratish
width = 100
depth = 100
x = np.linspace(0, width, width)
y = np.linspace(0, depth, depth)
X, Y = np.meshgrid(x, y)

# Sun'iy gravimetrik amplitudani yaratish
amplitude = np.sin(X/10) * np.cos(Y/10)

# 2. 2D Grafik: Aerogeozik Signalning Ta'siri
plt.figure(figsize=(8, 6))
plt.contourf(X, Y, amplitude, cmap='PuBu')
plt.colorbar(label='Amplitude')
plt.title('Aerogeozik Signal Amplitudasi')
plt.xlabel('Kenglik (m)')
plt.ylabel('Chuqurlik (m)')
plt.show()

# 3. 3D Grafik: Aerogeozik Signalning Taqsimlanishi
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, amplitude, cmap='viridis')
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=5)
ax.set_title('Aerogeozik Signal Taqsimlanishi (3D)')
ax.set_xlabel('Kenglik (m)')
ax.set_ylabel('Chuqurlik (m)')
ax.set_zlabel('Amplitude')
plt.show()

# 4. Animatsiya: Signalning O'zgarishini Ko'rsatish
fig, ax = plt.subplots(figsize=(8, 6))
im = ax.imshow(amplitude, cmap='YlGn', interpolation='nearest')
cbar = fig.colorbar(im)
cbar.set_label('Amplitude')

def update(frame):
    im.set_array(np.sin(X/10 + frame) * np.cos(Y/10 + frame))
    return [im]

ani = animation.FuncAnimation(fig, update, frames=range(100), interval=50, blit=True)
plt.show()

# 5. Konlarning Joylashuvi: Amplitudani yuqori bo'lgan hududlar
concentration = amplitude > 0.5
plt.figure(figsize=(8, 6))
plt.contourf(X, Y, concentration, cmap='coolwarm')
plt.colorbar(label='Concentration (Konlar)')
plt.title('Yerosti Konlarini Aniqlash')
plt.xlabel('Kenglik (m)')
plt.ylabel('Chuqurlik (m)')
plt.show()

```
