```python
# Barcha kerakli kutubxonalarni import qilish
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.animation as animation

# 1-qadam: Gravitatsion anomaliya ma'lumotlarini yaratish
x = np.linspace(0, 100, 200)  # X koordinatalar
z1 = 10 * np.exp(-((x - 30)**2) / 100)  # 1-qatlam anomaliyasi
z2 = -5 * np.exp(-((x - 70)**2) / 150)  # 2-qatlam anomaliyasi
total_anomaly = z1 + z2 + np.random.normal(0, 0.5, len(x))  # Shovqin qo'shish

# Ma'lumotlarni saqlash
data = pd.DataFrame({'X': x, 'Anomaly': total_anomaly})

# 2-qadam: 2D grafik yaratish
plt.figure(figsize=(10, 6))
plt.plot(data['X'], data['Anomaly'], label='Gravitatsion anomaliya', color='blue')
plt.title('Gravitatsion Anomaliyalar Grafigi')
plt.xlabel('X koordinata')
plt.ylabel('Anomaliya (mGal)')
plt.legend()
plt.grid(True)
plt.show()

# 3-qadam: 3D grafik yaratish
y = np.linspace(0, 100, 200)
X, Y = np.meshgrid(data['X'], y)
Z = np.exp(-((X - 50)**2 + (Y - 50)**2) / 300)  # Sun'iy 3D anomaliya

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_title('Gravitatsion Anomaliya (3D)')
ax.set_xlabel('X koordinata')
ax.set_ylabel('Y koordinata')
ax.set_zlabel('Anomaliya')
plt.show()

# 4-qadam: Qatlam chegaralarini aniqlash
threshold = 2  # Anomaliyaning keskin o‘zgarish chegarasi
boundaries = data[np.abs(data['Anomaly'].diff()) > threshold]

plt.figure(figsize=(10, 6))
plt.plot(data['X'], data['Anomaly'], label='Gravitatsion anomaliya', color='blue')
plt.scatter(boundaries['X'], boundaries['Anomaly'], color='red', label='Qatlam chegaralari')
plt.title('Qatlam chegaralarini aniqlash')
plt.xlabel('X koordinata')
plt.ylabel('Anomaliya (mGal)')
plt.legend()
plt.grid(True)
plt.show()

# 5-qadam: Animatsiya yaratish
# fig, ax = plt.subplots(figsize=(10, 6))
# line, = ax.plot([], [], lw=2)

# # Animatsiya boshlang'ich holati
# def init():
#     ax.set_xlim(0, 100)
#     ax.set_ylim(-15, 15)
#     return line,

# # Har bir ramkani yangilash funksiyasi
# def update(frame):
#     y = data['Anomaly'] + frame * 0.1  # Anomaliyaning vaqt bo‘yicha o‘zgarishi
#     line.set_data(data['X'], y)
#     return line,

# ani = animation.FuncAnimation(fig, update, frames=50, init_func=init, blit=True)
# plt.show()
```
