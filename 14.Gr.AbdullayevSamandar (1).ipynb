```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Gravitatsion doimiylik
G = 6.674 * 10**-11  # N·m²/kg²

# Parametrlarni aniqlash
mass1 = 5000  # Massa 1 (kg)
mass2 = 5000  # Massa 2 (kg)
distances = np.linspace(10, 200, 50)  # Masofalar (m)

# Og'irlik kuchi hisoblash funksiyasi
def gravity_force(m1, m2, r):
    return G * (m1 * m2) / r**2

# Og'irlik kuchi qiymatlarini hisoblash
forces = np.array([gravity_force(mass1, mass2, r) for r in distances])

# Gradientlarni hisoblash
gradients = np.gradient(forces, distances)

# 2D grafik: Og'irlik kuchi va gradientlari
plt.figure(figsize=(8, 6))
plt.plot(distances, forces, label="Og'irlik kuchi", color="blue")
plt.plot(distances, gradients, label="Gradientlar", color="red")
plt.xlabel("Masofa (m)")
plt.ylabel("Qiymatlar")
plt.title("Og'irlik kuchi va gradientlari")
plt.legend()
plt.grid(True)
plt.show()

# 3D grafik: Fazoviy og'irlik kuchi taqsimoti
x = np.linspace(-100, 100, 30)
y = np.linspace(-100, 100, 30)
X, Y = np.meshgrid(x, y)
Z = gravity_force(mass1, mass2, np.sqrt(X**2 + Y**2 + 200**2))

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap="viridis")
ax.set_xlabel("X (m)")
ax.set_ylabel("Y (m)")
ax.set_zlabel("Og'irlik kuchi (N)")
plt.title("Og'irlik kuchining 3D taqsimoti")
plt.show()
```
