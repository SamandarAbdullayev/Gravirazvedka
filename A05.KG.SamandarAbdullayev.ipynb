```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.animation as animation

# Sun'iy gravimetrik ma'lumotlarni yaratish
x = np.linspace(-10, 10, 100)  # X o'qi (masofa)
y = np.linspace(-10, 10, 100)  # Y o'qi (masofa)
X, Y = np.meshgrid(x, y)       # 2D tarmoq yaratish

# Yerosti ob'ektlari (masalan, zichroq qatlamlar) bilan gravimetrik anomaliya
Z = np.exp(-0.1 * (X**2 + Y**2))  # Sun'iy gravitatsion anomaliya
# Gravitatsion anomaliya hisoblash (masalan, Zichlikdan)
def grav_anomaliya(x, y):
    return np.exp(-0.1 * (x**2 + y**2))  # Sun'iy gravitatsion anomaliya (masalan, zichlik asosida)

# Gravitatsion anomaliyani hisoblash
anomaliya = grav_anomaliya(X, Y)
# 2D grafik
plt.figure(figsize=(8, 6))
plt.contourf(X, Y, anomaliya, 20, cmap='inferno')
plt.colorbar(label='Gravitatsion Anomaliya')
plt.title('Gravitatsion Anomaliya - 2D')
plt.xlabel('X (masofa)')
plt.ylabel('Y (masofa)')
plt.show()
```

```python
# 3D grafik
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# 3D grafikka gravitatsion anomaliya
ax.plot_surface(X, Y, anomaliya, cmap='BuPu', alpha=0.9 )

ax.set_title('Gravitatsion Anomaliya - 3D')
ax.set_xlabel('X (masofa)')
ax.set_ylabel('Y (masofa)')
ax.set_zlabel('Gravitatsion Anomaliya')
plt.show()
```
