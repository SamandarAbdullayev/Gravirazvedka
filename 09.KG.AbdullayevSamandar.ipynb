```python
import numpy as np
import matplotlib.pyplot as plt

# Gravitatsion doimiysi
G = 6.67430e-11

# Funksiyalar: gorizontal silindir uchun gravitatsion anomaliya va togri burchakli parallelepiped uchun gravitatsion anomaliya
def grav_cylinder(x, rho, r, d):
    return (2 * G * rho * r**2 * d) / (x**2 + d**2)

def grav_parallelopiped(x, rho, a, b, h):
    return (2 * G * rho * a**b * h) / (x**2 + h**2)

# Parametrlar
rho_cylinder = 2500  # kg/m^3 siln zichligi
r = 5
d = 20  # chuqurligi
rho_parallelopiped = 2500  # zichlig
a, b, h = 10, 5, 20  # p olchamlari

# Kuzatuv nuqtalari
x = np.linspace(-100, 100, 500)

# Gravitatsion anomaliyalar
g_cylinder = grav_cylinder(x, rho_cylinder, r, d)
g_parallelepiped = grav_parallelopiped(x, rho_parallelopiped, a, b, h)

# Grafiklar
plt.subplot(1, 2, 1)
plt.plot(x, g_cylinder, label="Gorizontal silinder")
plt.title("Gorizontal silinder uchun gravitatsion anomaliya", pad=30)
plt.xlabel("Masofa (m)")
plt.ylabel("Gravitatsion anomaliya (m/s^2)")
plt.grid(True)

# P uchun grafik
plt.subplot(1, 2, 2)
plt.plot(x, g_parallelepiped, label="Togri burchakli parallelopiped", color='orange')
plt.title("Togri burchakli p uchun gravitatsion anomaliya", pad=30)
plt.xlabel("Masofa (m)")
plt.ylabel("G anomaliya (m/s^2)")

# Grafiklarni korsatish
plt.tight_layout()
plt.show()
```
