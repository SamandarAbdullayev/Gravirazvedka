```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from mpl_toolkits.mplot3d import Axes3D  # 3D grafiklar uchun kutubxona

# 1. Olingan ma'lumotlar (simplified version)
# Balandlik (m), va gravitatsion kuch (mGal)
altitudes = np.array([0, 500, 1000, 1500, 2000, 2500])  # Balandlik
gravity_measurements = np.array([9.81, 9.79, 9.77, 9.75, 9.74, 9.73])  # Gravitatsiya kuchi (m/s^2)

# 2. Gravitatsion anomaliyalarning hisoblanishi
# Anomaliyaning hisoblanishi uchun sodda model
anomalia = gravity_measurements - np.mean(gravity_measurements)

# 3. 2D Grafik: Gravitatsion anomaliyaning balandlik bo'yicha o'zgarishi
plt.figure(figsize=(10, 6))
plt.plot(altitudes, anomalia, marker='o', color='b', linestyle='-', label='Gravitatsion Anomaliya')
plt.title('Gravitatsion Anomaliya va Balandlik o\'rtasidagi Bog\'liqlik')
plt.xlabel('Balandlik (m)')
plt.ylabel('Gravitatsion Anomaliya (mGal)')
plt.grid(True)
plt.legend()
plt.show()

# 4. 3D Grafik: Gravitatsion kuch va tuzilmalar o'rtasidagi bog'liqlik
# X: X o'qi (masofa), Y: Y o'qi (balandlik), Z: Z o'qi (anomaliya)

x = np.linspace(-5, 5, 10)  # X o'qi
y = np.linspace(0, 5000, 10)  # Balandlik
X, Y = np.meshgrid(x, y)
Z = np.sin(X) * np.cos(Y)  # 3D anomaliya (model sifatida)

# Matplotlib yordamida 3D grafik
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# 3D grafikka ma'lumotlarni joylashtirish
ax.plot_surface(X, Y, Z, cmap='viridis')

# Grafikni sozlash
ax.set_title('Yer Osti Tuzilmalari va Gravitatsion Kuchlar')
ax.set_xlabel('X (masofa)')
ax.set_ylabel('Balandlik (m)')
ax.set_zlabel('Gravitatsion Anomaliya (mGal)')

plt.show()

# # 5. Animatsiya yaratish: Vaqt o'zgarishiga qarab anomaliya o'zgarishini ko'rsatish
# fig, ax = plt.subplots(figsize=(10, 6))
# line, = ax.plot([], [], 'r-', lw=2)

# ax.set_xlim(0, 2500)
# ax.set_ylim(-0.02, 0.02)
# ax.set_xlabel('Balandlik (m)')
# ax.set_ylabel('Gravitatsion Anomaliya (mGal)')
# ax.set_title('Gravitatsion Anomaliya va Vaqt o\'rtasidagi Bog\'liqlik')

# # Funksiya: Har bir frame uchun anomaliya qiymatini yangilash
# def init():
#     line.set_data([], [])
#     return (line,)

# def animate(i):
#     ydata = np.sin(altitudes / 500) * np.exp(-i / 100)  # Animatsiyaning vaqt bo'yicha o'zgarishi
#     line.set_data(altitudes, ydata)
#     return (line,)

# # Animatsiya sozlamalari
# ani = FuncAnimation(fig, animate, frames=100, init_func=init, blit=True, interval=50)
# plt.show()
```
