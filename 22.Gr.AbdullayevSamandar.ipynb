```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# 1. Simulyatsiya qilingan ma'lumotlar tayyorlash
x = np.linspace(0, 10, 100)  # X o'qi (masofa)
y = np.linspace(0, 10, 100)  # Y o'qi (masofa)
X, Y = np.meshgrid(x, y)

# Gravitatsion kuchning simulyatsiyasi (masalan, g = 9.81 - x^2 - y^2)
G = 9.81 - X**2 - Y**2  # Gravitatsion kuchning qiymatlari

# 2. Gradientlarni hisoblash (markaziy farq metodi)
h = x[1] - x[0]  # X va Y o'qidagi farq
Gx, Gy = np.gradient(G, h, axis=(1, 0))  # Gorizontal gradientlar

# 3. 2D Grafik: Gravitatsion kuchning gorizontal gradientini ko'rsatish
plt.figure(figsize=(10, 6))
plt.contourf(X, Y, Gx, levels=50, cmap='viridis')
plt.colorbar(label='Gravitatsion kuch gradienti (mGal/m)')
plt.title('Gravitatsion Kuchning Gorizontal Gradienti (2D)')
plt.xlabel('X (masofa)')
plt.ylabel('Y (masofa)')
plt.show()

# 4. 3D Grafik: Gorizontal gradientlarni 3D shaklida ko'rsatish
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Gx, cmap='viridis')

ax.set_title('Gravitatsion Kuchning Gorizontal Gradienti (3D)')
ax.set_xlabel('X (masofa)')
ax.set_ylabel('Y (masofa)')
ax.set_zlabel('Gorizontal Gradient')
plt.show()


```
