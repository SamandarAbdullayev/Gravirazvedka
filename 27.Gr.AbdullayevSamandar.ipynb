```python
# 1. Kutubxonalarni import qilish
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.animation import FuncAnimation

# 2. Ma'lumotlar tayyorlash
Vg = np.array([10, 20, 30, 40, 50])  # G‘ovak hajmi (sm³)
Vj = np.array([100, 200, 300, 400, 500])  # Jinsning umumiy hajmi (sm³)

# 3. G‘ovaklar hajmini jinsning umumiy hajmiga nisbati (foiz)ni hisoblash
phi = (Vg / Vj) * 100  # Formula: ϕ = (Vg / Vj) * 100%

# 4. 2D graﬁk: G‘ovaklar hajmini jinsning umumiy hajmiga nisbati
plt.figure(figsize=(8, 6))
plt.plot(Vg, phi, marker='o', linestyle='-', color='b', label="ϕ = Vg / Vj × 100%")
plt.title("G‘ovaklar Hajmining Jinsning Umumiy Hajmiga Nisbati")
plt.xlabel("G‘ovak Hajmi (Vg), sm³")
plt.ylabel("Nisbati (ϕ), %")
plt.grid(True)
plt.legend()
plt.show()

# 5. 3D graﬁk: G‘ovak hajmi va jinsning umumiy hajmini ko‘rsatuvchi graﬁk
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# 3D nuqtalar
ax.scatter(Vg, Vj, phi, color='r')

# 3D graﬁk sozlash
ax.set_xlabel('G‘ovak Hajmi (Vg), sm³')
ax.set_ylabel('Jinsning Umumiy Hajmi (Vj), sm³')
ax.set_zlabel('Nisbati (ϕ), %')
ax.set_title("G‘ovak Hajmi va Jinsning Umumiy Hajmi")
plt.show()

# 6. Animatsiya: G‘ovak hajmining vaqt davomida o‘zgarishi
fig, ax = plt.subplots(figsize=(8, 6))
line, = ax.plot([], [], 'b-', lw=2)
ax.set_xlim(0, 50)
ax.set_ylim(0, 100)
ax.set_title("G‘ovak Hajmining O‘zgarishi (Animatsiya)")
ax.set_xlabel("G‘ovak Hajmi (Vg), sm³")
ax.set_ylabel("Nisbati (ϕ), %")

# Animatsiya uchun yangilash funksiyasi
def update(frame):
    Vg_current = np.linspace(10, frame, 50)  # Vg o'zgarishi
    phi_current = (Vg_current / 100) * 100  # Nisbati hisoblash
    line.set_data(Vg_current, phi_current)
    return line,

# Animatsiya yaratish
ani = FuncAnimation(fig, update, frames=np.linspace(10, 50, 50), interval=100, blit=True)
plt.show()

# 7. Natijalarni tahlil qilish
print("\nTahlil:")
print(f"1. G‘ovaklar hajmining jinsning umumiy hajmiga nisbati (ϕ) g‘ovak hajmi oshgani sayin ortadi.")
print(f"2. Nisbati o‘zgarishi jinsning umumiy hajmiga nisbatan aniq tendensiya ko‘rsatadi, ya'ni Vg oshganda ϕ ham oshadi.")
```
