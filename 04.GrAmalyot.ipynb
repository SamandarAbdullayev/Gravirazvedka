```python
# Og'irlik kuchining nisbiy o'lchovlarini aniqlash 1
import numpy as np
import matplotlib.pyplot as plt

# 2. Og'irlik kuchi uchun olchovlar tayyorlash
mass = np.array([10, 20, 30, 40, 50])  #massa (kg)
g = 9.81 

#3.Og'irlik kuchini hisoblash
weight = mass * g 

#4. nisbiy olchovlarni aniqlash
relative_weight = weight / weight.max()  

#5. Natijaviy o'lchovlarni aniqlash
print("Massa (kg):",mass)
print("og'irlik (N):", weight)
print("Nisbiy og'irlik kuchi:", relative_weight)

# 6 Grafik chizish
plt.figure(figsize=(10, 6))
plt.bar(mass, relative_weight, color='skyblue')
plt.title("Og'irlik Kuchining Nisbiy O'lchovlari")
plt.xlabel('Massa (kg)')
plt.ylabel("Nisbiy Og'irlik kuchi")
plt.xticks(mass)
plt.grid()
plt.show()

```

```python
# Og'irlik kuchining nisbiy o'lchovlarini aniqlash 2
# 1. Kutubxonalarni import qilish
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

#2. Tog' jinslari va qatlamlar
layers = {
    "Tosh": {"thickness": 20, "density":2.0},
    "Siltdtone": {"thickness": 30, "density":2.3},
    "Clay": {"thickness": 20, "density":2.0},
    "Sandstone": {"thickness": 40, "density":2.4},
}

# 3. Qatlamlar ma'lumotlarini hisoblash
depths = [0]
weights = []
for layer in layers.values():
    depths.append(depths[-1] + layer["thickness"])
    weight = layer["density"] * 9.81 * layer["thickness"]
    weights.append(weight)

#4. 2D vizualizatsiya
plt.figure(figsize=(12, 8))
plt.barh(list(layers.keys()), [layer["thickness"] for layer in layers.values()], color='lightblue')
plt.title("Yer Ostidagi Tog' Jinslari Qatlamlari")
plt.xlabel('Chuqurlik (m)')
plt.ylabel("Qatlamlar")
plt.grid()
plt.show()

#5. 3D vizualizatsiya
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

#Qatlamlarni chizish
for i, (layer_name, layer_info) in enumerate(layers.items()):
    ax.bar3d(0, i, depths[i], 1, 1, layer_info["thickness"], alpha=0.7, label=layer_name)

#Grafik ko'rsatkichlar
ax.set_title("Yer Ostidagi Tog' Jinslari va Geologik Qatlamlar")
ax.set_xlabel("X")
ax.set_ylabel('Qatlamlar')
ax.set_zlabel('Chuqurlik (m)')
ax.set_yticks(range(len(layers)))
ax.set_yticklabels(layers.keys())
plt.show()

# 6. Natijalarni korsatish
results_df = pd.DataFrame({
    'Qatlam': layers.keys(),
    'Zichlik (g/cm^3)':[layer['density']for layer in layers.values()],
    "Og'irlik Kuch (N/m^2)": weights
})
print("Natijalar:")
print(results_df)
```

```python
# Og'irlik kuchining nisbiy o'lchovlarini aniqlash 3 qism
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


#1 Parametrlarni aniqlash
depths = np.array([0, 50, 100, 150, 200]) # Chuqurlik
densities = np.array([2.5, 2.7, 2.8, 3.0, 3.2]) #zichlik

#2.Ogirlik kuchuni hisoblash
g = 9.81
gravity_anomalies = g * densities / 1000

# DAtaFrame yaratish
df = pd.DataFrame({
    'Chuqurlik (m)': depths,
    'Zichlik (g/cm^3)': densities,
    "O'irlik kuchi (m/s^2)": gravity_anomalies
})
#4. Natijalarni korsatish
print(df)

#5.Grafik chizish
plt.figure(figsize=(10, 6))
plt.plot(depths, gravity_anomalies, marker='o', linestyle='-', color='b')
plt.title("Chuqurlik va og'irlik kuchi o'rtasidagi bog'lanish")
plt.xlabel('Chuqurlik')
plt.ylabel("Og'irlik kuchi (m/s^2)")
plt.grid()
plt.show()
```

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt

# 1. Qatlam ma'lumotlari 
layers = {
    'Qum': {'density': 2.0, 'depth_range': (0, 50)},
    'Loy': {'density': 1.8, 'depth_range': (50, 150)},
    'Granit': {'density': 2.7, 'depth_range': (150, 200)},
}
# 2. Qatlamlar bo'yicha og'irlik kuchini hisoblash
g = 9.81 
depths = []
densities = []
gravity_anomalies = []
for layer, props in layers.items():
    depth_start, depth_end = props['depth_range']
    density = props['density']

    for depth in np.linspace(depth_start, depth_end, num=100):
        depths.append(depth)
        densities.append(density)
        gravity_anomalies.append(g * density) # og'irlik kuchi hisoblanadi
        
# 3. DataFrame yaratish
df = pd.DataFrame({
    'Chuqurlik (m)': depths,
    'Zichlik (g/cm^3)':densities,
    'Og\'irlik kuchi (m/s^2)': gravity_anomalies
})


# 4. Natijalrni ko'rsatish
print(df)

# Grtafik chizish 
plt.figure(figsize=(12, 6))
for layer, props in layers.items():
    plt.fill_betweenx(
        [props['depth_range'][0], props['depth_range'][1]],
        min(gravity_anomalies),
        max(gravity_anomalies),
        label=layer,
        alpha=0.5
    )
plt.title('Og\'irlik kuchining qatlamlariga ta\'siri')
plt.xlabel('Og\'irlik kuchi (m/s^2)')
plt.ylabel('Chuqurlik (m)')
plt.gca().invert_yaxis() # chuqurlikni  yuqoridan pastga qarab ko'rsatish
plt.legend()
plt.grid()
plt.show()
```
