```python
import numpy as np
import matplotlib.pyplot as plt

# Qatlamlar va ularning parametrlarini yaratish
layers = {
    'Layer 1': {'density': 2.5, 'depth': 0.5},
    'Layer 2': {'density': 2.7, 'depth': 1.0},
    'Layer 3': {'density': 3.0, 'depth': 1.5},
    'Layer 4': {'density': 3.3, 'depth': 2.0}
}

# Gravitatsion doimiy (G)
G = 6.67430e-11  # N⋅m²/kg²

# Qatlamlar uchun gravitatsion anomaliyalarning hisoblanishi
depths = []
anomali_values = []
layers_3d = []

# Qatlamlarni ishlash
for layer, properties in layers.items():
    density_change = properties['density'] - 2.5  # Zichlik o'zgarishi (odatiy qiymatdan farq)
    depth = properties['depth']
    anomaly = G * density_change * depth  # Gravitatsion anomaliya formulasi
    depths.append(depth)
    anomali_values.append(anomaly)
    layers_3d.append(layer)

# 3D graﬁk yaratish
depths_3d = np.array(depths)
anomalies_3d = np.array(anomali_values)

# Qatlamlarni raqamli indekslar bilan almashtirish
layer_indices = np.arange(len(layers_3d))

fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(depths_3d, layer_indices, anomalies_3d, color='r')

ax.set_xlabel("Qatlam Chuqurligi (m)")
ax.set_ylabel("Qatlamlar (Indeks)")
ax.set_zlabel("Gravitatsion Anomaliya (m/s²)")
ax.set_title("3D Gravitatsion Anomaliya Graﬁgi")

# Y o'qi bo'yicha qatlam nomlarini ko'rsatish
ax.set_yticks(layer_indices)
ax.set_yticklabels(layers_3d)

plt.show()
```
