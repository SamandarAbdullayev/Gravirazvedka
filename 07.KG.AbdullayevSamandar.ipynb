```python
import numpy as np
import matplotlib.pyplot as plt

# 2D maydonni yaratish
width = 100  # Kenglik
depth = 100   # Chuqurlik

# X, Y va Z koordinatalarini yaratish
x = np.linspace(0, width, width)
y = np.linspace(0, depth, depth)
X, Y = np.meshgrid(x, y)

# Sun'iy elektromagnit o'tkazuvchanlikni yaratish (tasodifiy qiymatlar)
conductivity = np.sin(X/10) * np.cos(Y/10)  # O'tkazuvchanlikning taqsimlanishi

# O'tkazuvchanlikni grafikda ko'rsatish
plt.figure(figsize=(8, 6))
plt.contourf(X, Y, conductivity, cmap='BuGn')
plt.colorbar(label='Conductivity')
plt.title('Electromagnetic Conductivity Distribution')
plt.xlabel('Width (m)')
plt.ylabel('Depth (m)')
plt.show()
```

```python
# 2D grafikda o'tkazuvchanlik va chuqurlik orasidagi bog'liqlikni ko'rsatish
plt.figure(figsize=(8, 6))
plt.plot(y, conductivity[:, 50], label='Conductivity at x=50')
plt.title('Conductivity vs Depth')
plt.xlabel('Depth (m)')
plt.ylabel('Conductivity')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

# 3D grafik yaratish
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# 3D yuzani yaratish
surf = ax.plot_surface(X, Y, conductivity, cmap='viridis')

# Rangli boshqaruv panelini qo'shish
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=5)

ax.set_title('Conductivity Distribution (3D)')
ax.set_xlabel('Width (m)')
ax.set_ylabel('Depth (m)')
ax.set_zlabel('Conductivity')
plt.show()
```

```python
import matplotlib.animation as animation

# Animatsiya funksiyasi
fig, ax = plt.subplots(figsize=(8, 6))
im = ax.imshow(conductivity, cmap='viridis', interpolation='nearest')

# Rangli boshqaruv panelini qo'shish
cbar = fig.colorbar(im)
cbar.set_label('Conductivity')

def update(frame):
    im.set_array(np.sin(X/10 + frame) * np.cos(Y/10 + frame))
    return [im]

ani = animation.FuncAnimation(fig, update, frames=range(100), interval=50, blit=True)
plt.show()
```

```python
# Rudalar joylashgan hududlarni aniqlash
rudalar = conductivity > 0.5  # O'tkazuvchanlik 0.5 dan katta bo'lgan joylar

# Rudalarni 3D grafikka qo'shish
fig_rudalar = plt.figure(figsize=(10, 8))
ax_rudalar = fig_rudalar.add_subplot(111, projection='3d')

# Rudalarni ko'rsatish
ax_rudalar.scatter(X[rudalar], Y[rudalar], conductivity[rudalar], c='r', marker='o', s=4, label="Rudalar")

ax_rudalar.set_title('Rudalar Joylashuvi')
ax_rudalar.set_xlabel('Kenglik (m)')
ax_rudalar.set_ylabel('Chuqurlik (m)')
ax_rudalar.set_zlabel('O\'tkazuvchanlik')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.animation as animation

# 1. EM Ma'lumotlarni Yaratish
width = 100
depth = 100
x = np.linspace(0, width, width)
y = np.linspace(0, depth, depth)
X, Y = np.meshgrid(x, y)

# Sun'iy elektromagnit o'tkazuvchanlik
conductivity = np.sin(X/10) * np.cos(Y/10)

# 2. 2D Grafik: O'tkazuvchanlik va Chuqurlik orasidagi bog'liqlik
plt.figure(figsize=(8, 6))
plt.plot(y, conductivity[:, 50], label='Conductivity at x=50')
plt.title('Conductivity vs Depth')
plt.xlabel('Depth (m)')
plt.ylabel('Conductivity')
plt.legend()
plt.grid(True)
plt.show()

# 3. 3D Grafik: O'tkazuvchanlikning taqsimlanishi
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, conductivity, cmap='viridis')
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=5)
ax.set_title('Conductivity Distribution (3D)')
ax.set_xlabel('Width (m)')
ax.set_ylabel('Depth (m)')
ax.set_zlabel('Conductivity')
plt.show()

# 4. Animatsiya: O'tkazuvchanlikdagi o'zgarishlarni ko'rsatish
fig, ax = plt.subplots(figsize=(8, 6))
im = ax.imshow(conductivity, cmap='viridis', interpolation='nearest')

cbar = fig.colorbar(im)
cbar.set_label('Conductivity')

def update(frame):
    im.set_array(np.sin(X/10 + frame) * np.cos(Y/10 + frame))
    return [im]

ani = animation.FuncAnimation(fig, update, frames=range(100), interval=50, blit=True)
plt.show()

# 5. Rudalar joylashuvi
rudalar = conductivity > 0.5
fig_rudalar = plt.figure(figsize=(10, 8))
ax_rudalar = fig_rudalar.add_subplot(111, projection='3d')
ax_rudalar.scatter(X[rudalar], Y[rudalar], conductivity[rudalar], c='r', marker='o', s=4, label="Rudalar")
ax_rudalar.set_title('Rudalar Joylashuvi')
ax_rudalar.set_xlabel('Kenglik (m)')
ax_rudalar.set_ylabel('Chuqurlik (m)')
ax_rudalar.set_zlabel('O\'tkazuvchanlik')
plt.legend()
plt.show()
```
