```python
# 1. Kutubxonalarni import qilish
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.animation import FuncAnimation

# 2. Ma'lumotlar tayyorlash
m = np.array([1, 2, 3, 4, 5])  # Massalar (kg)
g = 9.81  # Gravitatsion tezlik (m/s²)

# 3. Og‘irlik kuchini hisoblash
F = m * g  # Og‘irlik kuchi (N)

# 4. 2D graﬁk: Massaning og‘irlik kuchiga ta'sirini ko‘rsatish
plt.figure(figsize=(8, 6))
plt.plot(m, F, marker='o', linestyle='-', color='r', label="F = m * g")
plt.title("Massaning Og‘irlik Kuchiga Ta'siri")
plt.xlabel("Massa (kg)")
plt.ylabel("Og‘irlik Kuch (N)")
plt.grid(True)
plt.legend()
plt.show()

# 5. 3D graﬁk: Massaning og‘irlik kuchi va gravitatsion tezlikni ko‘rsatish
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# 3D nuqtalar
ax.scatter(m, F, g, color='b')

# 3D graﬁk sozlash
ax.set_xlabel('Massa (kg)')
ax.set_ylabel('Og‘irlik Kuch (N)')
ax.set_zlabel('Gravitatsion Tezlik (m/s²)')
ax.set_title("Massaning Og‘irlik Kuchiga Va Gravitatsion Tezlikka Ta'siri")
plt.show()

# 6. Animatsiya: Og‘irlik kuchining massaga bog‘liqligini ko‘rsatish
fig, ax = plt.subplots(figsize=(8, 6))
line, = ax.plot([], [], 'g-', lw=2)
ax.set_xlim(0, 5)
ax.set_ylim(0, 50)
ax.set_title("Og‘irlik Kuchining Massa O‘zgarishiga Bog‘liqligi")
ax.set_xlabel("Massa (kg)")
ax.set_ylabel("Og‘irlik Kuch (N)")

# Animatsiya uchun yangilash funksiyasi
def update(frame):
    m_current = np.linspace(1, frame, 50)  # Massa o‘zgarishi
    F_current = m_current * g  # Og‘irlik kuchi hisoblash
    line.set_data(m_current, F_current)
    return line,

# Animatsiya yaratish
ani = FuncAnimation(fig, update, frames=np.linspace(1, 5, 50), interval=100, blit=True)
plt.show()

# 7. Natijalarni tahlil qilish
print("\nTahlil:")
print(f"1. Og‘irlik kuchi massa oshgan sari ortadi. Bu, massa va og‘irlik kuchi o‘rtasidagi to‘g‘ri proporsional munosabatni anglatadi.")
print(f"2. Gravitatsion tezlik (g) Yer yuzasida doimiy bo‘lib, bu jarayonda unga hech qanday ta'sir ko‘rsatilmaydi.")
```
