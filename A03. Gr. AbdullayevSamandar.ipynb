```python
#Samandar
import numpy as np
import matplotlib.pyplot as plt

# Gravitatsion doimiysi
G = 6.67430e-11 # m^3 kg^-1 s^-2

# Yer parametrlari
M_earth = 5.972e24 # kg, yer massasi
R_earth = 6.371e6 # m, yer radiusi

# Og'irlik kuchi hisoblash funktsiyasi (balandlik bo'yicha o'zgarish)
def gravitational_acceleration(height):
    g = G * M_earth / (R_earth + height)**2
    return g

# Parametrlar
heights = np.linspace(0, 2000, 50) # 0 m dan 2000 m gacha balandliklar

# Yerning sirti va balandlikda og'irlik kuchini hisoblash
g_values = gravitational_acceleration(heights)

# Grafik chizish
plt.figure(figsize=(8, 6))
plt.plot(heights, g_values, label="Og'irlik kuchi", color='orange')
plt.xlabel('Balandlik (m)')
plt.ylabel("Og'irlik kuchi (m/s^2)")
plt.title("Og'irlik kuchining balandlik bo'yicha o'zgarishi")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Gravitatsion doimiysi
G = 6.67430e-11 # m^3 kg^-1 s^-2

# Massa xossalari 
masses = [
    {'rho': 5000, 'radius': 150, 'center': np.array([100, 200, -150])},
    {'rho': 3000, 'radius': 120, 'center': np.array([-300, 400, 150])},
    {'rho': 4500, 'radius': 180, 'center': np.array([50, -250, 50])},
]

# 2D maydonda kuzatuv (z=0
x = np.linspace(-500, 500, 100)
y = np.linspace(-500, 500, 100)
X, Y = np.meshgrid(x, y)
Z = np.zeros_like(X)

# Umumiy tortishish tezlashuvi komponentlari uchun massivlarni ishga tushirish
g_x_total = np.zeros_like(X)
g_y_total = np.zeros_like(X)

#  Har bir massadan tortishish kuchini hisoblash
for mass in masses:
    rho = mass['rho']
    r = mass['radius']
    center = mass['center']


    # Har bir jismning hajmi va massasini hisoblash (sferik massalarni hisobga olgan holda). 
    volume = (4/3) * np.pi * r**3
    M = rho * volume
    
    #  kuzatish nuqtalaridan massa markazigacha bo'lgan masofa komponentlarini hisoblash
    dx = X - center[0]
    dy = Y - center[1]
    dz = Z - center[2]

    # massa markazigacha bo'lgan massa 
    distance = np.sqrt(dx**2 + dy**2 + dz**2)

     # Gravitatsiyon tezlanish komponentlari
    g_x = G * M * dx / distance**3
    g_y = G * M * dy / distance**3

    # Umumiy tortishish maydoniga hissa qo'shing
    g_x_total += g_x 
    g_y_total += g_y


 # Umumiy tortishish maydonining kattaligi
g_total = np.sqrt(g_x_total**2 + g_y_total**2)

# grafika  chizish 
plt.figure(figsize=(14, 6))

# gravitatsion kuch vektorlari
plt.subplot(1, 2, 1)
plt.quiver(X, Y, g_x_total, g_y_total, g_total, scale=1e-6,cmap='plasma')
plt.title('Gravitatsion kuch chiziqlari')
plt.xlabel('X (m)')
plt.ylabel('Y (m)')
plt.colorbar(label='g magnitude (m/s^2)')
plt.grid(True)

# gravitatsion  potentsial kontur xaritasi
plt.subplot(1, 2, 2)
plt.contourf(X, Y, g_total, levels=20, cmap='inferno')
plt.title('Gravitatsion  potentsial kontur xaritasi')
plt.xlabel('X (m)')
plt.ylabel('Y (m)')
plt.colorbar(label='g magnitude (m/s^2)')
plt.grid(True)

plt.tight_layout()
plt.show()



    
```
