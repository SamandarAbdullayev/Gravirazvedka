```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
# Gravitatsion doimiysi
G = 6.67430e-11 #m^3 kg^-1 s^-2
# Gorizontal silindr uchun gravitatsion anomaliya
def grav_cylinder(x, rho, r, d):
    return (2 * G * rho * r**2 *d) /(x**2 + d**2)
rho_cylinder = 3500 # kg/m^3, silindirnig zichligi
r = 100 # m, silindirning radiusi
d = 100 # m, silindirning chuqurligi

# kuzatuv nuqtalari 
x = np.linspace(-200, 200, 100)
# Gravitatsion anomaliyalr 
g_cylinder = grav_cylinder(x, rho_cylinder, r, d, )
# grafiklar
plt.figure(figsize=(12,6))
# silindr uchun grafik 
plt.subplot(1, 2, 1)
plt.plot(x, g_cylinder, label="Gorizontal silindr", color='red')
plt.title("gorizontal silindr uchun gravitatsion anomaliya ", pad=30)
plt.xlabel("Masofa (m)")
plt.ylabel("Gravitatsion anomaliya (m/s^2")
plt.grid(True)

# 3D Grafik qurish

# Ma'lumotlarni yaratish
true_rho_cylinder = 3500  # kg/m^3
true_r_cylinder = 100  # m
true_d_cylinder = 100  # m
def plot_3d_objects():
    fig = plt.figure(figsize=(6, 10))
    ax1 = fig.add_subplot(111, projection='3d')
    
    # Gorizontal silindr
    u = np.linspace(0, 2 * np.pi, 100)
    v = np.linspace(-true_d_cylinder / 2, true_d_cylinder / 2, 10)
    U, V = np.meshgrid(u, v)
    X = true_r_cylinder * np.cos(U)
    Y = true_r_cylinder * np.sin(U)
    Z = V
    ax1.plot_surface(X, Y, Z, color='orange', alpha=0.9)
    ax1.set_title("Gorizontal Silindr")
    ax1.set_xlabel("X o'qi (m)")
    ax1.set_ylabel("Y o'qi (m)")
    ax1.set_zlabel("Z o'qi (m)")

    plt.tight_layout()
    plt.show()

plot_3d_objects()

# Grafiklarni ko'rsatish
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Parametrlar
G = 6.67430e-11  # Gravitasion doimiy (m^3 * kg^-1 * s^-2)
rho = 3000  # Zichlik (kg/m^3)
h = 200  # Qatlam chuqurligi (m)

# Masofalar
x = np.linspace(-1000, 1000, 500)  # Masofa (m)
y = np.linspace(-1000, 1000, 500)
X, Y = np.meshgrid(x, y)
r = np.sqrt(X**2 + Y**2 + h**2)  # Masofa hisoblash (Pifagor teoremasi)

# Gravitatsion anomaliya formulasi
delta_g = (2 * G * rho * h) / (r**2)

# 2D Grafik (Kontur)
plt.figure(figsize=(10, 6))
plt.contourf(X, Y, delta_g, 50, cmap='winter')
plt.colorbar(label='Gravitatsion Anomaliya (m/s²)')
plt.title('Gravitatsion Anomaliya - 2D Kontur grafigi')
plt.xlabel('X (m)')
plt.ylabel('Y (m)')
plt.show()

# 2D Grafik (Oddiy Line Grafik)
plt.figure(figsize=(10, 6))
# Oddiy chiziqli grafik uchun faqat x o'qi bo'yicha yasalgan anomaliyaning qiymatini olish
plt.plot(x, delta_g[int(len(x) / 2), :], label='Gravitatsion Anomaliya (X o`qi bo`ylab)', color='orange')
plt.title('Gravitatsion Anomaliya - 2D Oddiy Grafik')
plt.xlabel('X (m)')
plt.ylabel('Gravitatsion Anomaliya (m/s²)')
plt.legend()
plt.grid(True)
plt.show()

# 3D ko'rinish
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, delta_g, cmap='YlGn')

ax.set_title('Gravitatsion Anomaliya - 3D Ko`rinishi')
ax.set_xlabel('X (m)')
ax.set_ylabel('Y (m)')
ax.set_zlabel('Gravitatsion Anomaliya (m/s²)')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Parametrlar
G = 6.67430e-11  # Gravitatsion doimiy (m³·kg⁻¹·s⁻²)
zichlik = 3000  # kg/m³
chuqurlik = 200  # m
radius = np.linspace(1, 500, 100)

# Animatsiya uchun funksiya
fig, ax = plt.subplots()
line, = ax.plot([], [], lw=2)
ax.set_xlim(0, 500)
ax.set_ylim(0, 2e-5)

def init():
    line.set_data([], [])
    return line,

def animate(i):
    delta_g = (4 * np.pi * G * zichlik * chuqurlik) / (radius ** 2 + i * 0.1)  # Vaqt o‘tishi bilan radiusning o‘zgarishi
    line.set_data(radius, delta_g)
    return line,

ani = FuncAnimation(fig, animate, init_func=init, frames=100, interval=50, blit=True)
plt.title("Gravitatsion Anomaliya Animatsiyasi")
plt.xlabel("Masofa (m)")
plt.ylabel("Gravitatsion Anomaliya (m/s²)")
plt.show()
```
