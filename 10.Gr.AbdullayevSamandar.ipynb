```python
# Kerakli kutubxonalarni import qilish
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.animation import FuncAnimation

# 1. Ma'lumotlarni tayyorlash: Qatlamlar uchun zichlik va chuqurlik qiymatlari
layers = ['Qum', 'Loy', 'Granit']
densities = [1800, 2000, 2700]  # Zichlik (kg/m^3)
depths = [50, 100, 350]  # Chuqurlik (m)
g = 9.81  # Yerning tortish kuchi (m/s^2)

# 2. Og‘irlik kuchini va anomaliyasini hisoblash
weight_forces = [density * g * depth for density, depth in zip(densities, depths)]
avg_weight_force = np.mean(weight_forces)  # O‘rtacha og‘irlik kuchi

# Anomaliyalari hisoblash: og‘irlik kuchidan o‘rtacha qiymatni ayirish
anomalies = [force - avg_weight_force for force in weight_forces]

# 3. Natijalarni DataFrame formatida saqlash
data = {
    'Qatlam': layers,
    'Zichlik (kg/m³)': densities,
    'Chuqurlik (m)': depths,
    'Og‘irlik kuchi (N)': weight_forces,
    'Anomaliya (N)': anomalies
}

df = pd.DataFrame(data)

# DataFrame ni chiqarish
print("Og‘irlik kuchining anomaliyasini hisoblash natijalari:")
print(df)

# 4. 2D grafikka og‘irlik kuchining anomaliyasini chizish
plt.figure(figsize=(8, 6))
plt.bar(df['Qatlam'], df['Anomaliya (N)'], color='skyblue')
plt.title('Og‘irlik kuchining anomaliyasi', fontsize=14)
plt.xlabel('Qatlamlar', fontsize=12)
plt.ylabel('Anomaliya (N)', fontsize=12)
plt.grid(True)
plt.show()

print("3D grafikka og‘irlik kuchining taqsimoti:")


# 3D grafikka og‘irlik kuchining taqsimotini chiziqlar bilan ko‘rsatish
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

x = np.array([1, 2, 3])  # Qatlamlar (soni)
y = np.array(depths)  # Chuqurliklar
z = np.array(anomalies)  # Anomaliyalar

# Chiziqlar bilan chizish
ax.plot(x, y, z, color='blue', marker='o')  # Chiziq va nuqta ko‘rinishi

# O'qlarni belgilash
ax.set_xlabel('Qatlamlar')
ax.set_ylabel('Chuqurlik (m)')
ax.set_zlabel('Og‘irlik kuchining anomaliyasi (N)')

# Grafikni ko‘rsatish
plt.show()




# 7. Natijalarni tahlil qilish
print("\nOg‘irlik kuchining anomaliyasi:")
print(df[['Qatlam', 'Anomaliya (N)']])

# Qaysi qatlam og‘irlik kuchining anomaliyasini ko‘proq ta’sir qiladi?
max_anomaly_layer = df.loc[df['Anomaliya (N)'].idxmax()]['Qatlam']
print(f"\nOg‘irlik kuchining anomaliyasini eng ko‘p ta’sir qilgan qatlam: {max_anomaly_layer}")
```
